# Работа с линейным графиком

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1, 5, 10, 15, 20]
y = [1, 7, 3, 5, 11]
```

```python
line = plt.plot(x, y)
plt.setp(line, linestyle='--')
```

```python
x = [1, 5, 10, 15, 20]
y1 = [1, 7, 3, 5, 11]
y2 = [i*1.2 + 1 for i in y1]
y3 = [i*1.2 + 1 for i in y2]
y4 = [i*1.2 + 1 for i in y3]
```

```python
plt.plot(x, y1, '-', x, y2, '--', x, y3, '-.', x, y4, ':')
```

```python
plt.plot(x, y, '--r')
```

```python
plt.plot(x, y, 'ro')
```

```python
plt.plot(x, y, 'bx')
```
